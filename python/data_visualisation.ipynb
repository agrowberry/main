```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import serial
```

```python
odo = pd.read_csv("Odo.csv")
acc = pd.read_csv("Accel.csv") 
```

```python
odo["X1"] = odo["X"]
odo["Y1"] = odo["Y"]
odo["Phi1"] = odo["Phi"]

odo = odo.drop(["X", "Y", "Phi"], axis=1)

for i in np.linspace(1, 9, 9):
    i = int(i)
    file = "Odo" + str(i) + ".csv"
    df = pd.read_csv(file)
    odo.loc[:, "X"+str(i+1)] = df["X"]
    odo.loc[:, "Y"+str(i+1)] = df["Y"]
    odo.loc[:, "Phi"+str(i+1)] = df["Phi"]


```

```python
rect_x = np.array([])
rect_y = np.array([])

rect_y = np.append(rect_y, np.linspace(-170, 170, 1000))
rect_y = np.append(rect_y, np.linspace(170, 170, 1000))
rect_y = np.append(rect_y, np.linspace(170, -170, 1000))
rect_y = np.append(rect_y, np.linspace(-170, -170, 1000))

rect_x = np.append(rect_x, np.linspace(270, 270, 1000))
rect_x = np.append(rect_x, np.linspace(270, -270, 1000))
rect_x = np.append(rect_x, np.linspace(-270, -270, 1000))
rect_x = np.append(rect_x, np.linspace(-270, 270, 1000))
```

```python
def argmin(x, y, rect_x=rect_x, rect_y=rect_y):
    distances = ((rect_x-x)**2 + (rect_y-y)**2)**0.5
    return min(distances)
```

```python
for index, row in odo.iterrows():
    for i in range(10):
        i = str(i+1)
        odo.loc[index, "Diff"+i] = argmin(row["X"+i], row["Y"+i])


   
```

```python
odo["MeanDiff"] = odo[["Diff"+str(i+1) for i in range(10)]].mean(axis=1)

odo
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=odo["Unnamed: 0"]+1, y=odo["MeanDiff"]))

fig.update_layout(template="plotly_white", width=800, height=600)

fig.update_layout(font=dict(family="Computer Modern", size=14),
                                     margin=go.layout.Margin(
                                        l=0, #left margin
                                        r=0, #right margin
                                        b=0, #bottom margin
                                        t=0, #top margin
                                    ))

fig.update_yaxes(title="Mean Boundary Error")

fig.update_xaxes(title="Collision Number", tick0=0, dtick=2)

fig.show()
```

```python
fig = go.Figure()

#54 x 34
fig.add_trace(go.Scatter(x=odo["X1"], y=odo["Y1"], mode="markers", marker=dict(size=10), showlegend=False))
fig.add_trace(go.Scatter(x=odo["X1"], y=odo["Y1"], mode="lines", line=dict(color="green", width=0.5), showlegend=False))

fig.update_layout(width=800, height=800, template="plotly_white")

fig.add_shape(type="rect",
    x0=-270, y0=-170, x1=270, y1=170 ,
    line=dict(color="black", width=2),
    fillcolor=None
)

for point, data in odo.iterrows():
    fig.add_annotation(go.layout.Annotation(dict(
                x = data["X1"] + 40*np.cos(data["Phi3"]),
                y = data["Y1"] + 40*np.sin(data["Phi3"]),
                xref="x", yref="y",
                text="",
                showarrow = True,
                axref = "x", ayref='y',
                ax = data["X1"],
                ay = data["Y1"],
                arrowhead = 3,
                arrowwidth = 1.5,
                arrowcolor = 'rgb(255,51,0)',)
            ))

fig.update_yaxes(
    scaleanchor = "x",
    scaleratio = 1
  )

fig.update_layout(font=dict(family="Computer Modern", size=14),
                                     margin=go.layout.Margin(
                                        l=0, #left margin
                                        r=0, #right margin
                                        b=0, #bottom margin
                                        t=0, #top margin
                                    ))

fig.update_yaxes(title="Y Displacement [mm]")

fig.update_xaxes(title="X Displacement [mm]")

fig.show()

fig.write_image("BoxTest1.pdf")
```
